```python
r,c = txt.shape
time = []
Azimuth = []
Altitude = []
```

```python
for i in range(r):
    row= str(txt.iloc[i]).split()
    
    time.append(row[1]+' '+row[2])
    Azimuth.append(float(row[-6]))
    Altitude.append(float(row[-5]))
    
```

```python
azu = np.array(Azimuth)
```

```python
azu = 360-azu
```

```python
len(Altitude)
```

```python
chiron = {
    'Time' : time,
    'Azimuth' : Azimuth,
    'Altitude' : Altitude
}
```

```python
ch = pd.DataFrame(chiron, columns = ['Time', 'Azimuth', 'Altitude'])
```

```python
ch.to_csv('chiron_altaz.csv', index=False)
```
