## planets

```python
from skyfield.api import load, wgs84, Angle
import pickle
import pandas as pd
ts= load.timescale()
```

```python
f1 = open('planet_position_with_time_18_20.pkl', 'rb')
p = pickle.load(f1)
```

```python
pp=p['dist_radec']
time_utc = list(p['time_p'])
len(time_utc)
time=[]
print(time_utc[0].utc_jpl())
for i in range(len(time_utc)):
    time.append(time_utc[i].utc_jpl())
```

```python
planets=['mercury', 'venus', 'mars', 'jupiter', 'saturn', 'uranus', 'neptune']
mer = pp['mercury']
ve = pp['venus']
ma = pp['mars']
ju = pp['jupiter']
sa = pp['saturn']
ur = pp['uranus']
ne = pp['neptune']
```

```python
len(mer['ra'].hours)
```

```python
# mercury
ra=list(mer['ra'].hours)
dec=list(mer['dec'].degrees)
dist=list(mer['distance'].au)
merr={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
mercury=pd.DataFrame(merr, columns=['Time','ra','dec','distance'])
```

```python
# venus
ra=list(ve['ra'].hours)
dec=list(ve['dec'].degrees)
dist=list(ve['distance'].au)
vee={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
venus=pd.DataFrame(vee, columns=['Time','ra','dec','distance'])
```

```python
# mars
ra=list(ma['ra'].hours)
dec=list(ma['dec'].degrees)
dist=list(ma['distance'].au)
maa={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
mars=pd.DataFrame(maa, columns=['Time','ra','dec','distance'])
```

```python
# jupiter
ra=list(ju['ra'].hours)
dec=list(ju['dec'].degrees)
dist=list(ju['distance'].au)
juu={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
jupiter=pd.DataFrame(juu, columns=['Time','ra','dec','distance'])
```

```python
# saturn
ra=list(sa['ra'].hours)
dec=list(sa['dec'].degrees)
dist=list(sa['distance'].au)
saa={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
saturn=pd.DataFrame(saa, columns=['Time','ra','dec','distance'])
```

```python
# uranus
ra=list(ur['ra'].hours)
dec=list(ur['dec'].degrees)
dist=list(ur['distance'].au)
urr={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
uranus=pd.DataFrame(urr, columns=['Time','ra','dec','distance'])
```

```python
# neptune
ra=list(ne['ra'].hours)
dec=list(ne['dec'].degrees)
dist=list(ne['distance'].au)
nee={'Time':time, 'ra':ra, 'dec':dec, 'distance':dist}
neptune=pd.DataFrame(nee, columns=['Time','ra','dec','distance'])
```

```python

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('major_planets_2020_22.xlsx', engine='xlsxwriter')

# Write each dataframe to a different worksheet.
mercury.to_excel(writer, sheet_name='mercury')
mars.to_excel(writer, sheet_name='mars')
jupiter.to_excel(writer, sheet_name='jupiter')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```

```python
mercury.to_csv('mercury.csv',index=False)
venus.to_csv('venus.csv',index=False)
mars.to_csv('mars.csv',index=False)
jupiter.to_csv('jupiter.csv',index=False)
saturn.to_csv('saturn.csv',index=False)
uranus.to_csv('uranus.csv',index=False)
neptune.to_csv('neptune.csv',index=False)
```
